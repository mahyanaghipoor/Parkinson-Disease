```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import ttest_ind
from statsmodels.stats.multitest import multipletests
import numpy as np

# تابع رسم Boxplot
def plot_boxplot(data, labels, title, file_path):
    plt.figure(figsize=(8, 6))
    plt.boxplot(data, labels=labels)
    plt.title(title)
    plt.ylabel('Value')
    plt.grid(True)
    plt.savefig(file_path)
    plt.close()

# خواندن داده‌ها
file_path = 'D:/DOC Aarabi/parkinson/new paper/new13-06/text/structuralGradients_hc_ses01.xlsx'
df = pd.read_excel(file_path)

# تقسیم گروه‌ها
healthy_group = df.iloc[:, 0:13]       # 1 تا 13
patient_group = df.iloc[:, 13:126]     # 14 تا 126
group_3 = df.iloc[:, 126:179]          # 127 تا 179 (گروه جدید 1)
group_4 = df.iloc[:, 179:]             # 180 تا آخر (گروه جدید 2)

# مسیر ذخیره تصاویر
output_paths = [
    'D:/DOC Aarabi/parkinson/new paper/DATA/text/boxplot_gradient_patient.png',
    'D:/DOC Aarabi/parkinson/new paper/DATA/text/boxplot_gradient_group3.png',
    'D:/DOC Aarabi/parkinson/new paper/DATA/text/boxplot_gradient_group4.png'
]

# لیست گروه‌های مقایسه → (نام گروه, داده, مسیر ذخیره)
comparison_groups = [
    ('Patient (14-126)', patient_group, output_paths[0]),
    ('Group3 (127-179)', group_3, output_paths[1]),
    ('Group4 (180-end)', group_4, output_paths[2])
]

# ایجاد فایل خروجی برای مقادیر آماری و p-value ها
with open('D:/DOC Aarabi/parkinson/new paper/DATA/text/statistics_output.txt', 'w') as f:
    f.write("Healthy Group (1-13) - Standard Deviation:\n")
    f.write(f"{healthy_group.std().to_string()}\n\n")

    for name, group_data, plot_path in comparison_groups:
        # آماره توصیفی
        f.write(f"{name} - Standard Deviation:\n{group_data.std().to_string()}\n\n")

        # آزمون t-test برای هر ستون جداگانه → Healthy vs ستون فعلی از گروه
        p_values = []
        for col in range(group_data.shape[1]):
            t_stat, p_val = ttest_ind(healthy_group.values.flatten(), group_data.iloc[:, col].values.flatten(), equal_var=False)
            p_values.append(p_val)

        # تصحیح FDR
        reject, pvals_corrected, _, _ = multipletests(p_values, method='fdr_bh')

        # ذخیره خروجی‌ها
        f.write(f"{name} - Raw p-values:\n{np.array(p_values)}\n\n")
        f.write(f"{name} - FDR-corrected p-values:\n{pvals_corrected}\n\n")
        f.write(f"{name} - Significant (after FDR):\n{reject}\n\n")

        # رسم Boxplot
        plot_boxplot(
            [healthy_group.values.flatten(), group_data.values.flatten()],
            ['Healthy', name],
            f"Healthy vs {name}",
            plot_path
        )


```
